```python
#1)Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 
#(keep in mind that the date format is YYYY-MM-DD).
year= requests.get('https://www.quandl.com/api/v3/datasets/FSE/EON_X?start_date=2017-01-01&end_date=2017-12-31&api_key=fAGNik47Q__1mgr8Z8Zn')
year.json()
```

```python
#2)Convert the returned JSON object into a Python dictionary
print(type(year.json()))
year_dict = year.json()
type(year_dict)
```

```python
#3)Calculate what the highest and lowest opening prices were for the stock in this period.
data =year_dict['dataset']['data']
print(data)

col_names = year_dict['dataset']['column_names']
print(col_names)
# Data type of 'data'
print(type(data))
#Data type of 'col_names
print(type(col_names))
```

```python
opening= []
for row in data:
    opening.append(row[1])
print(opening)
```

```python
#4)What was the largest change in any one day (based on High and Low price)?
high_price= []
for row in data:
    high_price.append(row[2])
print(high_price)
print(type(high_price))
```

```python
low_price = []
for row in data:
    low_price.append(row[3])
print(low_price)
print(type(low_price))
```

```python
#list not tuples see below for operation after tuple coupleing
import operator
list(map(operator.sub, high_price, low_price))
```

```python
high_low_pairs = zip(high_price, low_price)
print(high_low_pairs)
import operator
list(map(operator.sub, high_price, low_price))
```

```python
#5)What was the largest change between any two days (based on Closing Price)?
closing_prices=[]
for row in data:
    closing_prices.append(row[4])
print(closing_prices)
```

```python
max(closing_prices)-min(closing_prices)
```

```python
#6)What was the average daily trading volume during this year?
trading_vol = []
for row in data:
    trading_vol.append(row[6])
print(trading_vol)
```

```python
sum(trading_vol)/len(trading_vol)
```
